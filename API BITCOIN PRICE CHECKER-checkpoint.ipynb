# MAYANK'S CRYPTO PRICE CHECKER PROJECT USING COINGECKO API

```python
import pandas as pd
import requests
from pycoingecko import CoinGeckoAPI
import matplotlib.pyplot as plt
```

```python
cg=CoinGeckoAPI()
```

```python
ohlc = cg.get_coin_ohlc_by_id(id = 'ethereum', vs_currency = 'usd', days = 30)
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency = 'inr', days = 30)
```

```python
bitcoin_data
```

```python
bitcoin_data['prices']
```

```python
df = pd.DataFrame(bitcoin_data)
```

```python
df
```

```python
data= pd.DataFrame(bitcoin_data['prices'],columns =['TimeStamp','Price'])
```

```python
data
```

```python
data['Date']=pd.to_datetime(data['TimeStamp'],unit='ms')
```

```python
data
```

```python
candlestick_data = data.groupby(data.Date.dt.date).agg({'Price': ['min', 'max', 'first', 'last']})
```

```python
candlestick_data
```

```python
import plotly.graph_objects as go
import plotly.offline as pyo

fig = go.Figure(data=[go.Candlestick(x=candlestick_data.index,
                                     open=candlestick_data['Price']['first'],
                                     high=candlestick_data['Price']['max'],
                                     low=candlestick_data['Price']['min'],
                                     close=candlestick_data['Price']['last'])])

fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title='Date', yaxis_title='Price(INR)', title='Bitcoin candlestick chart over the past 30 days')

# Save the plot to an HTML file
pyo.plot(fig, filename='bitcoin_candlestick_graph.html')
```
